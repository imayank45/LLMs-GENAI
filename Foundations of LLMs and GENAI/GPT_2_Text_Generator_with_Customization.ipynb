```python
from transformers import pipeline
```

```python
# load GPT-2 model
generator = pipeline('text-generation', model='gpt2')
```

```python
# text generation function
def generate_text(prompt, max_length=100, temperature=0.7, top_k=50):
  output = generator(
      prompt,
      max_length=max_length,
      temperature=temperature,
      top_k=top_k
  )
  return output[0]['generated_text']
```

```python
# User Interaction Loop
print("📝 Welcome to GPT-2 Text Generator! ")
print("--------------------------------------------------")

while True:
    user_prompt = input("\n💬 Enter your prompt (or type 'exit' to quit): ")
    if user_prompt.lower() == "exit":
        print("👋 Exiting the text generator. Goodbye!")
        break

    # Temperature and top-k settings
    try:
        max_length = int(input("📏 Enter maximum text length (e.g., 100): ") or 100)
        temperature = float(input("🔥 Enter temperature (0.1 - 1.0) [default: 0.7]: ") or 0.7)
        top_k = int(input("🎯 Enter top-k sampling value (e.g., 50): ") or 50)
    except ValueError:
        print("❗ Invalid input. Using default settings.")
        max_length, temperature, top_k = 100, 0.7, 50

    # Generate Text
    print("\n✨ Generated Text:")
    print("--------------------------------------------------")
    print(generate_text(user_prompt, max_length, temperature, top_k))
    print("--------------------------------------------------")
```
